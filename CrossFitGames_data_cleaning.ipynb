```python
import pandas as pd 
```

```python
data = pd.read_excel('CrossFit_Games_Athletes_Men_2019.xlsx')
```

```python
data.head(10)
```

### Removing the first 9 null rows

```python
data_copy = data.iloc[9:]

data_copy = data_copy.reset_index(drop = True )

data_copy.head()
```

### Promoting the first row as the header

```python
data_copy.columns = data_copy.iloc[0].values
data_copy = data_copy.iloc[1:].reset_index(drop=True )
data_copy.head()
```

```python
data_copy.shape
```

### Renaming the column names for consistency

```python
data_copy= data_copy.rename(columns ={'countryoforigin':'country_of_origin','competitorid':'competitor_id','countryoforigincode':'country_code',
                                      'overallrank':'overall_rank','countryoforiginname':'country_name','bibid':'bib_id',
                                      'competitorname':'competitor_name','Number of Records':'no_of_records','overallscore':'overall_score','affiliatename':'affiliate_name',
                                     'affiliateid':'affiliate_id'})
data_copy.head()
```

```python
data_copy.isnull().sum()
```

### Typecasting 

```python
data_copy['affiliate_name'] = data_copy['affiliate_name'].fillna('unknown')
data_copy['affiliate_id'] = data_copy['affiliate_id'].fillna('0').astype(int)
```

```python
data_copy['overall_rank'] = data_copy['overall_rank'].replace(r'(\D+)','',regex =True).astype(int)
data_copy['overall_score'] = data_copy['overall_score'].fillna(data_copy['overall_score'].mean())
```

```python

data_copy[['height','weight','no_of_records','competitor_id','bib_id','age','overall_score','division','gender']] = data_copy[['height','weight','no_of_records','competitor_id','bib_id','age','overall_score','division','gender']]\
.astype({'height':'float','weight':'float','no_of_records':'int','competitor_id':'int','bib_id':'int','age':'int','overall_score':'int','division': 'category','gender': 'category'})
```

```python

data_copy =data_copy.dropna(axis = 1)
data_copy = data_copy.drop(['profilepics3key'],axis = 1)
data_copy.head()
```

```python
data_copy.dtypes
```

```python
data_copy.describe()
```

```python
mens_data = data_copy.copy()
```
